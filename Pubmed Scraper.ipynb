```python
import csv
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import undetected_chromedriver as uc

# Path to your ChromeDriver
PATH = r"C:\Program Files (x86)\chromedriver.exe"

# Create a service object for ChromeDriver
service = Service(executable_path=PATH)

# Set up Chrome options
options = webdriver.ChromeOptions()
options.add_argument('--disable-blink-features=AutomationControlled')
options.add_argument("--incognito")  # Enables incognito mode
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option("useAutomationExtension", False)

driver = uc.Chrome()

# URL to scrape
url = "https://pubmed.ncbi.nlm.nih.gov/?term=improve+employee+wellness"
driver.get(url)

# Create or open the CSV file to store the data
with open('titles_and_authors.csv', 'w', newline='', encoding='utf-8') as csvfile:
    # Create a CSV writer object
    csvwriter = csv.writer(csvfile)
    
    # Write the header row
    csvwriter.writerow(['Title', 'Authors'])
    
    # Scrape the first 5 pages
    for j in range(5):
        # Wait for the page to fully load
        time.sleep(3)  # Adjust this time if needed
        
        # Find all the article titles on the page
        entries = driver.find_elements(By.XPATH, "//a[@class='docsum-title']")
        
        # Find all the corresponding author elements
        author_elements = driver.find_elements(By.XPATH, "//span[@class='docsum-authors full-authors']")
        
        for i, entry in enumerate(entries):
            try:
                # Get the title
                title = entry.text
                
                # Get the corresponding authors
                authors = author_elements[i].text if i < len(author_elements) else "N/A"
                
                # Print title and authors (for debugging)
                print(f"Title: {title}")
                print(f"Authors: {authors}")
                
                # Write title and authors to the CSV file
                csvwriter.writerow([title, authors])
            except Exception as e:
                print(f"Error extracting data for one entry: {e}")
        
        # Find and click the 'Next' button to go to the next page
        try:
            next_button = driver.find_element(By.XPATH, "//button[@class='button-wrapper next-page-btn']")
            next_button.click()
        except Exception as e:
            print(f"Error clicking the next button: {e}")
            break

# Close the browser after scraping
driver.quit()

print("Scraping completed and data saved to titles_and_authors.csv")
```
