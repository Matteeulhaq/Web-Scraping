```python


import time
import csv
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import undetected_chromedriver as uc

# Path to your ChromeDriver
PATH = r"C:\Program Files (x86)\chromedriver.exe"

# Create a service object for ChromeDriver
service = Service(executable_path=PATH)

# Set up Chrome options
options = webdriver.ChromeOptions()
options.add_argument('--disable-blink-features=AutomationControlled')
options.add_argument("--incognito")  # Enables incognito mode
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option("useAutomationExtension", False)

# Initialize undetected ChromeDriver
driver = uc.Chrome()

# URL to scrape
url = "https://www.academia.edu/search?&q=Employee%20wellness&utf8=%E2%9C%93"
driver.get(url)

# Open a CSV file to store the data
with open('Academia.csv', 'w', newline='', encoding='utf-8') as csvfile:
    # Create a CSV writer object
    csvwriter = csv.writer(csvfile)
    
    # Write the header row
    csvwriter.writerow(['Title', 'Authors', 'Link'])

    # Scrape the first 5 pages
    for j in range(5):
        time.sleep(3)  # Adjust this time if needed

        # Find all the entries on the current page
        entries = driver.find_elements(By.XPATH, "//div[@class='work-card-wrapper search_work_card_redesign-WorkCardWrapper-cls2-2o3N search_work_card_redesign-WorkCardWrapper-cls1-2lIn']")
        
        for entry in entries:
            try:
                # Extract title, link, and authors for each entry
                title_entry = entry.find_element(By.XPATH, ".//a[@class='search_work_card_redesign-WorkTitle-cls2-1nCY search_work_card_redesign-WorkTitle-cls1-2OpX']")
                title = title_entry.text
                link = title_entry.get_attribute("href")
                
                author_element = entry.find_element(By.XPATH, ".//div[@class='search_work_card_redesign-AuthorNameTitle-cls2-1L34 search_work_card_redesign-AuthorNameTitle-cls1-hs9R']")
                authors = author_element.text
                
                # Write the title, authors, and link to the CSV file
                csvwriter.writerow([title, authors, link])
                
                # Print the data for debugging
                print(f"Title: {title}")
                print(f"Authors: {authors}")
                print(f"Link: {link}\n")
                
            except Exception as e:
                print(f"Error extracting data for one entry: {e}")
        
        # Find and click the 'Next' button to go to the next page
        try:
            if j == 0:
                next_button = driver.find_element(By.XPATH, "//button[@class='Search-PageButton-cls2-3MJN Search-PageButton-cls1-2e75']")
            else:
                next_button = driver.find_element(By.XPATH, '/html/body/div[8]/div[1]/div[1]/div[4]/div/div[2]/main/div[4]/button[2]')
            next_button.click()
        except Exception as e:
            print(f"Error clicking the next button: {e}")
            break

# Close the browser after scraping
driver.quit()

print("Scraping completed and data saved to Academia.csv")
```
